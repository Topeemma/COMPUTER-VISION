# COMPUTER VISION
**Computer vision is a field of artificial intelligence that gives computers the ability to see, understand, and interpret the visual world
The goal of computer vision is to enable computers to see the world in the same way that humans do. This is a challenging task, as the visual world is complex and constantly changing. However, recent advances in artificial intelligence and machine learning have made significant progress in the field of computer vision**

**what is an Image**
> a digital image is not a continuous image, like a painting or a photograph. Instead, it is made up of a grid of pixels, and each pixel has a specific value that represents its color or brightness. It is an array of pixels arranged in columns and rows. Pixels are the elements of an image that contain information about intensity and color.Pixels are arranged in the form of a matrix. This is known as an RGB image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Create a blank canvas filled with zeros
canvas2 = np.zeros((1000, 1000, 3), dtype=np.uint8)

# Draw a rectangle on the canvas
cv2.rectangle(canvas2, pt1=(200, 200), pt2=(800, 800), color=(50, 0, 200), thickness=-1)

# Display the canvas with the rectangle
plt.imshow(canvas2)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Create a blank canvas filled with zeros
canvas3 = np.zeros((1000, 1000, 3), dtype=np.uint8)

# Draw circles on the canvas
cv2.circle(canvas3, center=(430, 500), radius=300, color=(120, 40, 60), thickness=8)
cv2.circle(canvas3, center=(250, 630), radius=150, color=(0, 140, 80), thickness=-1)

# Display the canvas with circles
plt.imshow(canvas3)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
cv2.circle(canvas3,center=(430,500),radius = 300,color=(120,40,60),thickness=8)
cv2.circle(canvas3,center=(250,630),radius = 150,color=(0,140,80),thickness=-1)
cv2.circle(canvas3,center=(200,350),radius = 150,color=(50,0,200),thickness=-1)

plt.imshow(canvas3)

```

```python
cav4 = np.zeros((1000,1000))

red = cav4.copy()
blue = cav4.copy()
green = cav4.copy()

red[200:400,200:400] = 225
blue[250:450,250:450] = 225
green[300:500,300:500] = 225

red = Image.fromarray(red).convert("L")
blue = Image.fromarray(blue).convert("L")
green = Image.fromarray(green).convert("L")

pic = Image.merge("RGB",[red,blue,green])
plt.imshow(pic)
```
